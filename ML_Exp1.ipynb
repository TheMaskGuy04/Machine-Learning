```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/drive/MyDrive/VIT Sem VII/ML/1. housing.csv')
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
data.isnull().sum()
```

```python
data.reset_index(drop=True, inplace=True)
data
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['ocean_proximity'] = le.fit_transform(data['ocean_proximity'])
data
```

```python
import matplotlib.pyplot as plt
# Check the correlation between features and target variable
correlation_matrix = data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
x = data.copy()

sns.scatterplot(x = x["total_bedrooms"], y = x["median_house_value"])
```

```python
x = x[x['total_bedrooms'] < 2800]
```

```python
sns.scatterplot(x = x["total_rooms"], y = x["median_house_value"])
```

```python
x = x[x['total_rooms'] < 15000]
```

```python
sns.scatterplot(x = x["population"], y = x["median_house_value"])
x = x[x['population'] < 6500]
```

```python
sns.scatterplot(x = x["households"], y = x["median_house_value"])
x = x[x['households'] < 2000]
```

```python
sns.scatterplot(x = x["median_income"], y = x["median_house_value"])
x = x[x['median_income'] < 9]
```

```python
x_data = x.drop("median_house_value" , axis = 1).values
y_data = x["median_house_value"].values
```

```python
from sklearn.preprocessing import PolynomialFeatures
feature = PolynomialFeatures(degree=3, include_bias=True, interaction_only=False)
x_data = feature.fit_transform(x_data)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler(copy=True, with_mean=True, with_std=True)
x_data = scaler.fit_transform(x_data)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train , y_test = train_test_split(x_data,y_data, test_size=0.25 , random_state=42)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
lr = LinearRegression()
```

```python
lr.fit(x_train , y_train)
```

```python
lr.score(x_train , y_train)
```

```python
lr.score(x_test , y_test)
```

```python
y_pred =lr.predict(x_test)
df = pd.DataFrame({"Y_test": y_test , "Y_pred" : y_pred})
```

```python
plt.figure(figsize=(10,8))
plt.plot(df[:50])
plt.legend(["Actual" , "Predicted"])
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error: {rmse}")
print(f"R-squared: {r2}")
```

```python
input_features = {
    "longitude": -121.09,
    "latitude": 39.48,
    "housing_median_age": 25,
    "total_rooms": 1665,
    "total_bedrooms": 374,
    "population": 845,
    "households": 330,
    "median_income": 1.5603,
    "ocean_proximity": 1
}

input_df = pd.DataFrame([input_features])
input_df_poly = feature.transform(input_df)
input_df_scaled = scaler.transform(input_df_poly)

prediction = lr.predict(input_df_scaled)

prediction = prediction[0]

print(f"Predicted median house value: {prediction}")
```
